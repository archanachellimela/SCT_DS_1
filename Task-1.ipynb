```python
import matplotlib.pyplot as plt

ages = [23, 45, 12, 67, 34, 45, 22, 34, 56, 44, 33]
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

genders = ["Male", "Female", "Male", "Female", "Female"]
counts = [genders.count("Male"), genders.count("Female")]

plt.bar(["Male", "Female"], counts, color=["blue", "pink"])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {"gender": ["Male","Female","Female","Male","Female","Male"]}
sns.countplot(x="gender", data=data)
plt.title("Gender Distribution")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"Gender": ["Male","Female","Female","Male","Male"]})
df["Gender"].value_counts().plot(kind="bar", color="green")
plt.title("Gender Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

group1 = np.random.normal(40, 10, 200)
group2 = np.random.normal(30, 5, 200)

plt.hist(group1, bins=20, alpha=0.5, label="Group 1")
plt.hist(group2, bins=20, alpha=0.5, label="Group 2")
plt.legend()
plt.title("Two Group Age Distribution")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ["A", "B", "C"]
men = [20, 34, 30]
women = [25, 32, 34]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, men, width, label="Men")
plt.bar(x + width/2, women, width, label="Women")
plt.legend()
plt.xticks(x, categories)
plt.title("Grouped Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [30, 25, 20, 25]
labels = ["Red", "Blue", "Green", "Yellow"]
colors = ["red", "blue", "green", "gold"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")
plt.title("Colored Pie Chart")
plt.show()
```
